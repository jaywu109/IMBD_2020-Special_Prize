```python
# 檢查缺失值的 function，列出有缺失值那列整列的資料以及那列總共有幾個缺失值
def Nan_Check(data):
    is_NaN = data.isnull()
    row_has_NaN = is_NaN.any(axis=1)
    nan_train=data[row_has_NaN]
    nan_stat=nan_train.isnull().sum(axis=1)
    return nan_train,nan_stat
```

```python
# 把每個方位資料變成 X 與 Y 兩欄，先把所有方位資料從 raw_trainX 分開出來為 o_cat_raw_trainX

o_cat_raw_trainX=raw_trainX.loc[:,raw_trainX.columns[128:152]]
cat_index=[]
old_index=raw_trainX.columns[128:152].array
for i in old_index:
    cat_index.append(i+'_Y')
    cat_index.append(i+'_X')
raw_trainX=raw_trainX.drop(raw_trainX.columns[128:152],axis=1)

# 先創建值為 0 的 Dataframe 方便將方位資料拆分填進去
n_cat_raw_trainX = pd.DataFrame(0, index=o_cat_raw_trainX.index, columns=cat_index)
```

```python
# 因為方位資料內有可能有 nan 會是福點數的格式，所以先看是不是純文字檔
# 若讀到非文字檔，先將那個 feacture 的 X&Y 欄都填入 nan

for index,row in o_cat_raw_trainX.iterrows():
    for column in old_index:
        if type(row[column]) == str:
            cat= row[column].split(';')
            if cat[0] =='N':
                n_cat_raw_trainX.loc[index,(column+'_Y')]=0
            if cat[0] =='U':
                n_cat_raw_trainX.loc[index,(column+'_Y')]=float(cat[1])
            if cat[0] =='D':
                n_cat_raw_trainX.loc[index,(column+'_Y')]= (-1*float(cat[1]))
            if cat[2] =='N':
                n_cat_raw_trainX.loc[index,(column+'_X')]=0
            if cat[2] =='R':
                n_cat_raw_trainX.loc[index,(column+'_X')]=float(cat[3])
            if cat[2] =='L':
                n_cat_raw_trainX.loc[index,(column+'_X')]=(-1*float(cat[3]))
        else:
            n_cat_raw_trainX.loc[index,(column+'_Y')]=np.nan
            n_cat_raw_trainX.loc[index,(column+'_X')]=np.nan
```

```python
# 把楚利後的方位資料接回去raw_trainX
raw_trainX=pd.concat([raw_trainX, n_cat_raw_trainX], axis=1)
```

```python
df=raw_trainX.drop(['Number'], axis=1)
```

採用 Scikit-learn IterativeImputer 進行缺失值預測填補

預設 Estimator 採用 BayesianRidge，之後可採用 KNN 或 RandomForest 等方式進行預測填補

[不同 Estimator 的實驗比較](https://scikit-learn.org/stable/auto_examples/impute/plot_iterative_imputer_variants_comparison)

[教學1](https://towardsdatascience.com/preprocessing-regression-imputation-of-missing-continuous-values-f612179bafb4)

[教學2](https://andymdc31.github.io/missing_data_and_using_sklearns_iterativeimputer)

```python
# 採用 Scikit-learn IterativeImputer 進行缺失值預測填補
imp = IterativeImputer(max_iter=10, verbose=0)
imp.fit(df)
imputed_df = imp.transform(df)
imputed_df = pd.DataFrame(imputed_df, index=o_cat_raw_trainX.index, columns=df.columns)
```

```python
# 檢查缺失值，發現為空序列代表沒缺失值了
_, stat= Nan_Check(imputed_df)
stat
```

```python
final_trainX=pd.concat([raw_trainX['Number'], imputed_df], axis=1)
```
